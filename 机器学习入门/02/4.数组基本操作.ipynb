## 1.4 生成随机数组

### 1.4.1 正态分布

```python
x1 = np.random.normal(1.75,1,100000000)
```

```python
x1
```

```python
plt.figure(figsize=(20,8),dpi=100)

plt.hist(x1,1000)

plt.show()
```

```python
stock_change = np.random.normal(0,1,(4,5))
stock_change
```

### 1.4.2 均匀分布

```python
x2 = np.random.uniform(-1,1,100000000)
```

```python
x2
```

```python
plt.figure(figsize=(20,8),dpi=100)

plt.hist(x2,1000)

plt.show()
```
