# 缺失值处理

```python
movie = pd.read_csv("./data/IMDB-Movie-Data.csv")
```

```python
movie.head()
```

## 缺失值是nan

```python
np.all(pd.notnull(movie)) # 里面如果有一个缺失值，就会返回False
```

```python
np.any(pd.isnull(movie)) # 里面如果有一个缺失值，就会返回True
```

```python
data = movie.dropna()
```

```python
np.all(pd.notnull(data))
```

```python
movie["Revenue (Millions)"].mean()
```

```python
movie["Revenue (Millions)"].fillna(movie["Revenue (Millions)"].mean(),inplace=True)
```

```python
movie.head()
```

```python
for i in movie.columns:
    if np.any(pd.isnull(movie[i])) == True:
        print(i)
        movie[i].fillna(movie[i].mean(),inplace=True)
```

```python
np.all(pd.notnull(movie))
```

# 数据离散化

```python
data = pd.read_csv("./data/stock_day.csv")
```

```python
data.head()
```

```python
p_change = data["p_change"]
```

```python
p_change.head()
```

```python
# 自动分成差不多数量的类别
qcut = pd.cut(p_change, 10)
qcut.value_counts()
```

```python
# 指定分组区间
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
p_count = pd.cut(p_change, bins)
```

```python
p_count.value_counts()
```

```python
dummies = pd.get_dummies(p_count, prefix="rise")
```

```python
dummies.head()
```

# 合并

```python
pd.concat([data, dummies],axis=1).head()
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
'key2': ['K0', 'K1', 'K0', 'K1'],
'A': ['A0', 'A1', 'A2', 'A3'],
'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
'key2': ['K0', 'K0', 'K0', 'K0'],
'C': ['C0', 'C1', 'C2', 'C3'],
'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

# 交叉表透视表

```python
data.head()
```

```python
data.index
```

```python
time = pd.to_datetime(data.index)
```

```python
time.weekday
```

```python
time.day
```

```python
data["week"] = time.weekday
```

```python
data.head()
```

```python
data["p_n"] = np.where(data["p_change"] > 0, 1, 0)
```

```python
data.head()
```

```python
count = pd.crosstab(data["week"], data["p_n"])
count
```

```python
sum = count.sum(axis=1).astype(np.float32)
sum
```

```python
ret = count.div(sum, axis=0)
```

```python
ret
```

```python
ret.plot(kind="bar", stacked=True)
plt.show()
```

```python
data.pivot_table(["p_n"], index="week")
```
