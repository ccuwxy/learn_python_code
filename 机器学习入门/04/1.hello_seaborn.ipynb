```python
import seaborn as sns
import numpy as np
import pandas as pd
```

# 绘制单变量分布

```python
np.random.seed(0)  # 确定随机数生成器生成的数是一样的

arr = np.random.randn(100)

sns.distplot(arr, bins=10, hist=True, kde=True, rug=True)
```

# 绘制双变量分布

## 绘制散点图

```python
df = pd.DataFrame({"x":np.random.randn(500),"y":np.random.randn(500)})
```

```python
df.head()
```

```python
sns.jointplot("x", "y", data=df, kind="scatter", color="r", size=10, ratio=5, space=0)
```

## 绘制二维直方图

```python
sns.jointplot("x", "y", data=df, kind="hex")
```

## 绘制核密度估计图形

```python
sns.jointplot("x", "y", data=df, kind="kde")
```

# 类别散点图

```python
data= sns.load_dataset("tips",data_home="seaborn-data-master", cache=True)
```

```python
data.head()
```

```python
sns.stripplot(x="day",y="total_bill",data = data)
```

```python
sns.stripplot(x="day",y="total_bill",data = data, hue="time")
```

```python
sns.stripplot(x="day",y="total_bill",data = data, hue="time", jitter=False)
```

```python
sns.swarmplot("day","total_bill", data=data)
```

# 类别内数据分布
## 箱型图

```python
sns.boxplot("day", "total_bill",data = data, hue="time", palette=["g","y"], saturation=0.5)
```

## 绘制提琴图

```python
sns.violinplot("day","total_bill",data=data)
```

# 类别内的估计统计

```python
sns.barplot("day","total_bill",data = data)
```

```python
sns.pointplot("day","total_bill",data=data)
```
